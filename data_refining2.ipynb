```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file='./MLD.csv'
data=pd.read_csv(file)
data.head(1)
```

```python
label_final=data['CHEST CIRCUMFERENCE(mm)']
feature=data.drop(['WEIGHT(N)','HEAD CIRCUMFERENCE(mm)','HAND LENGTH(mm)','CHEST CIRCUMFERENCE(mm)','AGE (months)'],axis=1)
```

```python
plt.plot(feature,label_final)
```

```python
feature.head(1)
```

```python
del data
label_final=np.array(label_final)
feature=np.array(feature)
```

```python
from six.moves import cPickle as pickle
import os
root='.'
pickle_file=os.path.join(root,'.pickle')
with open(pickle_file,'wb') as f:
    save={
    'label_final':label_final,
    'features':feature,
    }
    pickle.dump(save,f,pickle.HIGHEST_PROTOCOL)
    f.close()
```
